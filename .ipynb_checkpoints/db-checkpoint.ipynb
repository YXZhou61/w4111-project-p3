```python
%%sql
DROP TABLE IF EXISTS reviews CASCADE;
create table reviews(
  restaurant_id text REFERENCES Restaurants,
  user_id text REFERENCES Users, 
  text text[],
  review_time timestamp[],
  review_rating stars[], 
  PRIMARY KEY(restaurant_id,user_id)
);
```

```python
%%sql
CREATE TYPE stars AS ENUM ('1.0','1.5','2.0','2.5','3.0','3.5','4.0','4.5','5.0');
CREATE TYPE plevel AS ENUM ('$','$$','$$$','$$$$');
```

```python
%%sql 
DROP TABLE IF EXISTS Restaurants CASCADE;
CREATE TABLE Restaurants (
    restaurant_id text,
    restaurant_name text,
    address_street text,
    restaurant_phone text,
    is_closed boolean,
    price_level plevel,
    rating stars,
    review_count integer,
    PRIMARY KEY(restaurant_id)
);
DROP TABLE IF EXISTS Users CASCADE;
CREATE TABLE Users(
  user_id text,
  user_name text,
  user_phone text,
  user_password text,
  PRIMARY KEY(user_id)
);

DROP TABLE IF EXISTS Regions CASCADE;
CREATE TABLE Regions(
  city text, 
  state text, 
  zip_code text, 
  PRIMARY KEY(zip_code)
);
DROP TABLE IF EXISTS Cuisines CASCADE;
CREATE TABLE Cuisines(
  cuisine_name text,
  PRIMARY KEY(cuisine_name)
);

DROP TABLE IF EXISTS Hours CASCADE;
create table  Hours(
  hour_id SERIAL,  
  monday text[], 
  tuesday text[], 
  wednesday text[], 
  thursday text[], 
  friday text[], 
  saturday text[], 
  sunday text[],
  PRIMARY KEY(hour_id)
);

DROP TABLE IF EXISTS locates CASCADE;
create table locates(
  zip_code text NOT NULL REFERENCES Regions,
  restaurant_id text REFERENCES Restaurants,
  PRIMARY KEY(restaurant_id)
);
DROP TABLE IF EXISTS serves CASCADE;
create table serves(
  cuisine_name text NOT NULL REFERENCES Cuisines,
  restaurant_id text REFERENCES Restaurants,
  PRIMARY KEY(restaurant_id)
);
DROP TABLE IF EXISTS reviews CASCADE;
create table reviews(
  restaurant_id text REFERENCES Restaurants,
  user_id text REFERENCES Users,
  text text[],
  review_time timestamp[],
  review_rating stars[], 
  PRIMARY KEY(restaurant_id,user_id)
);
DROP TABLE IF EXISTS recommends CASCADE;
create table recommends(
  restaurant_id text REFERENCES Restaurants,
  user_id text REFERENCES Users,
  PRIMARY KEY(restaurant_id,user_id)
);
DROP TABLE IF EXISTS eats CASCADE;
create table eats(
  restaurant_id text REFERENCES Restaurants,
  user_id text REFERENCES Users,
  when_ timestamp[],
  PRIMARY KEY(restaurant_id,user_id)
);
DROP TABLE IF EXISTS reserves CASCADE;
create table reserves(
  restaurant_id text REFERENCES Restaurants,
  user_id text REFERENCES Users,
  reservation_time timestamp[], 
  number_of_people int[], 
  primary key (restaurant_id,user_id)
);
DROP TABLE IF EXISTS opens_at CASCADE;
create table opens_at(
  restaurant_id text REFERENCES Restaurants,
  hour_id serial NOT NULL REFERENCES Hours,
  primary key (restaurant_id)
)
```

```python
from flask_login import UserMixin
from typing import Dict, Optional
class User(UserMixin):
    def __init__(self, id: str, name: str, phone: str, password: str):
        self.id = id
        self.name = name
        self.phone = phone
        self.password = password

    @staticmethod
    def get(user_id: str) -> Optional["User"]:
        return users.get(user_id)

    def __str__(self) -> str:
        return f"<Id: {self.id}, Username: {self.name}, Phone: {self.phone}>"

    def __repr__(self) -> str:
        return self.__str__()

```

```python
def insertIntoTable(df, table,serial=False):
    """
    Using cursor.executemany() to insert the dataframe
    """
    # Create a list of tuples from the dataframe values
    tuples = list(set([tuple(x) for x in df.to_numpy()]))

    # Comma-separated dataframe columns
    #cols = ','.join(list(df.columns))
    col_values = ','.join(['%%s']*len(df.columns))

    #When parameters are used, in order to include a literal % in the query you can use the %% string:
    # SQL query to execute
    #When parameters are used, in order to include a literal % in the query you can use the %% string:
    #query = "INSERT INTO %s(%s) VALUES(%%s,%%s,%%s,%%s)" % (table, cols)
    if serial:
        col_names = ','.join((list(df.columns)))
        query = "INSERT INTO %s(%s) VALUES({})".format(col_values) % ('Hours',col_names)
    else:
        query = "INSERT INTO %s VALUES({})".format(col_values) % (table)

    try:
        cur.executemany(query, tuples)
        conn.commit()

    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        conn.rollback()
        return 1
```

```python
data_folder = 'D:/Yuxuan Zhou/Columbia/courses/COMS W4111 Introduction to Databases/project/data/'
```

```python
import time
import pandas as pd
conv = lambda x: x.replace('[','{').replace(']','}') if x!='' else None
conv2 = lambda x: None if type(x)==float else x
conv3 = lambda x: '{'+x+'}'
converters = {
              'monday':conv,
              'tuesday':conv,
              'wednesday':conv,
              'thursday':conv,
              'friday':conv,
              'saturday':conv,
              'sunday':conv}
              #'price':conv2}
cuisines = ['Chinese','American','Indian','Italian','Japanese']
cities = ['New York','San Francisco','Los Angeles','Chicago','Boston']
Hour_table = pd.read_csv(data_folder+'Hours.csv',index_col=0,converters=converters)
insertIntoTable(Hour_table.drop(columns = 'id',axis=1), 'Hours',True)

Cui_table = pd.DataFrame({'cuisines':cuisines})
insertIntoTable(Cui_table, 'Cuisines')

for city in cities:
    for cuisine in cuisines:
        print(city,cuisine)
        temp = pd.read_csv('{}{}-{}.csv'.format(data_folder,cuisine,city),index_col=0)
        print(temp)
        temp['price'] = temp['price'].apply(conv2)
        temp['rating'] = temp['rating'].apply(str)
        Res_table = temp[['id', 'name', 'street', 'phone',
                  'is_closed', 'price', 'rating', 'review_count']]
        insertIntoTable(Res_table, 'Restaurants')

        Reg_table = temp[['city', 'state', 'zip_code']]
        
        loc_table = temp[['zip_code','id']]
        ser_table = temp[['cuisine','id']]
        
        opens_table = pd.merge(temp,Hour_table,on=['monday','tuesday','wednesday','thursday','friday','saturday','sunday'],suffixes=['_res','_hour'])[['id_res','id_hour']]
        insertIntoTable(Reg_table, 'Regions')
        time.sleep(2)
        insertIntoTable(loc_table, 'locates')
        insertIntoTable(ser_table, 'serves')
        insertIntoTable(opens_table, 'opens_at')

res_table = pd.read_csv(data_folder+'reservation.csv',index_col=0,converters={'reservation_time':conv3,'num_of_people':conv3})
eat_table = pd.read_csv(data_folder+'eat.csv',index_col=0,converters={'when':conv3})
User_table = pd.read_csv(data_folder+'user.csv',index_col=0)
rec_table = pd.read_csv(data_folder+'recommendation.csv',index_col=0)
rev_table = pd.read_csv(data_folder+'reviews.csv',index_col=0)
insertIntoTable(User_table, 'Users')
insertIntoTable(res_table, 'reserves')
insertIntoTable(eat_table, 'eats')
insertIntoTable(rec_table, 'recommends')
insertIntoTable(rev_table, 'reviews')

```
